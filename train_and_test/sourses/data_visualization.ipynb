```python
import pandas as pd
import numpy as np
import re
import matplotlib.pyplot as plt
import seaborn as sns
plt.style.use('seaborn')
# import paths as path
```

```python
train_path = '../../dataset/Train.csv'
test_path = '../../dataset/Test.csv'
base_path = '../../dataset/'
save_path = '../../outputs/data_visualization/'
```

```python
train_df = pd.read_csv(train_path)
train_df.head()
```

```python
train_df.Width.values.min(), train_df.Width.values.max()
```

```python
train_df.Height.values.min(), train_df.Height.values.max()
```

```python
fig = plt.figure(figsize=(12,8))
fig.patch.set_facecolor('white')
sns.histplot(data=train_df, x='ClassId', bins=43)
plt.title('Traffic sign images in the dataset for training')
plt.xticks(range(0,43))
plt.savefig(save_path+'train_dataset_distribution.jpg');
```

```python
img_path_array = np.array(train_df.Path)
np.random.shuffle(img_path_array)
fig = plt.figure(figsize=(10,10))
fig.patch.set_facecolor('white')
plt.title('Traffic sign images in the dataset for training')
for i in range(25):
    img = plt.imread(base_path + img_path_array[i])
    
    plt.subplot(5,5,i+1)
    plt.xticks([])
    plt.yticks([])
    plt.grid(False)
    plt.imshow(img, cmap=plt.cm.binary)
    plt.xlabel(re.findall(r'\d{1,2}',img_path_array[i])[0])
plt.savefig(save_path+'train_img.jpg')
plt.show()
```
